```python
# task-1

class Calculator:
  def __init__(self,fn,o,ln):
    self.fn = fn
    self.o = o
    self.ln = ln


  
  def add(self):
    res = self.fn + self.ln
    print('Result:',res)

  def subtract(self):
    res = self.fn - self.ln
    print('Result:',res)

  def multiply(self):
    res = self.fn * self.ln
    print('Result:',res)

  def divide(self):
    res = self.fn / self.ln
    print('Result:',res)


print('Let’s Calculate!')
f = int(input('Value 1: '))
o = input('Operator: ')
l = int(input('Value 2: '))

c1 = Calculator(f, o, l)

if o == '+':
  c1.add()

elif o == '-':
  c1.subtract()

elif o == '*':
  c1.multiply()

elif o == '/':
  c1.divide()
```

```python
# task-5

class Shape:
  def __init__(self,sn,fv,sv):
    self.sn = sn
    self.fv = fv
    self.sv = sv

  def area(self):
    if self.sn == "Triangle":
      print('Area: ',0.5*self.fv*self.sv)
    elif self.sn == "Square":
      print('Area: ',self.fv*self.sv)
    elif self.sn == "Rhombus":
      print('Area: ',0.5*self.fv*self.sv)
    elif self.sn == "Rectangle":
      print('Area: ',self.fv*self.sv)
    else:
      print('Area: Shape unknown')

            

# Write your code here
triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
# task-6
class Calculator:
  def __init__(self):
    print('Calculator is ready!')
    self.val1 = 0
    self.val2 = 0
    self.sign = '+'
    
  def calculate(self,val1,val2,sign):
    self.val1 = val1
    self.val2 = val2
    self.sign = sign

    if self.sign == '+':
      self.result = self.val1 + self.val2
    elif self.sign == '-':
      self.result = self.val1 - self.val2
    elif self.sign == '*':
      self.result = self.val1 * self.val2
    elif self.sign == '/':
      self.result = self.val1 / self.val2
    return self.result 

  def showCalculation(self):
    print(f'{self.val1}{self.sign}{self.val2} = {self.result}')

# Write your code here

c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
# task-7

class Student:
  def __init__(self,n,id,dp,cg):
    self.n = n
    self.id = id
    self.dp = dp
    self.cg = cg
    self.cgpa = 0

  def calculate_CGPA(self):
    l = len(self.cg)
    socg = 0
    for i in range(l):
      a = self.cg[i]*3
      socg += a
    self.cgpa = socg/(l*3)

  def print_details(self):
    print(f'Name: {self.n}, ID: {self.id}')
    print('Department: ',self.dp)
    print('CGPA: ',self.cgpa)
    if self.cgpa>3.8:
      print("Your academic standing is 'Highest Distinction'")
    elif 3.65<self.cgpa<=3.8:
      print("Your academic standing is 'High Distinction'")
    elif 3.50<self.cgpa<=3.65:
      print("Your academic standing is 'Distinction'")
    elif 2.00<self.cgpa<=3.50:
      print("Your academic standing is 'Satisfactory'")
    elif self.cgpa<2.00:
      print('Sorry, you can’t Graduate')




# Write your code here
s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
print("==========================")
s1.print_details()
print("==========================")
s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
print("==========================")
s2.print_details()
print("==========================")
s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
print("==========================")
s3.print_details()
```

```python
# task-8

class Shinobi:
  def __init__(self,n,r):
    self.name = n
    self.rank = r
    self.salary = 0
    self.mission = 0

  def changeRank(self,cr):
    self.rank = cr

  def calSalary(self,n):
    self.mission = n

    if self.rank == 'Genin':
      self.salary = self.mission*50
    elif self.rank == 'Chunin':
      self.salary = self.mission*100
    else:
      self.salary = self.mission*500

  def printInfo(self):
    print('Name: ',self.name)
    print('Rank: ',self.rank)
    print('Number of mission: ',self.mission)
    print('Salary: ',self.salary)


# Write your code here.
naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```

```python
# task-10

class UberEats:
  def __init__(self,na,nu,lo):
    self.na = na
    self.nu = nu
    self.lo = lo
    self.ofd = {}
    self.t = 0
    print(f'{self.na}, welcome to UberEats!')

  def add_items(self,ffn,sfn,ffa,sfa):
    self.ffn = ffn
    self.sfn = sfn
    self.ffa = ffa
    self.sfa = sfa

    self.ofd[ffn] = ffa
    self.ofd[sfn] = sfa

    self.t = self.ffa + self.sfa

  def print_order_detail(self):
    return f"User details: Name: {self.na}, Phone: {self.nu}, Address: {self.lo}, Orders: {self.ofd}\n Total Paid Amount: {self.t}"


# Write your code here
order1 = UberEats("Shakib", "01719658xxx", "Mohakhali")
print("=========================")
order1.add_items("Burger", "Coca Cola", 220, 50)
print("=========================")
print(order1.print_order_detail())
print("=========================")
order2 = UberEats ("Siam", "01719659xxx", "Uttara")
print("=========================")
order2.add_items("Pineapple", "Dairy Milk", 80, 70)
print("=========================")
print(order2.print_order_detail())
```
