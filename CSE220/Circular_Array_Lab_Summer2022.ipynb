```python
class CircularArray:
  def __init__(self, lin, st, sz):
    # Initializing Variables
    self.start = st
    self.size = sz
    self.cir = [None]*len(lin)
    
    # if lin = [10, 20, 30, 40, None]
    # then, CircularArray(lin, 2, 4) will generate
    # cir = [40, null, 10, 20, 30]

    idx = self.start
    for i in range(self.size):
      self.cir[idx] = lin[i]
      idx = (idx + 1)%len(self.cir)
    #print(self.cir)
    
    # To Do. 
    # Hints: set the values for initialized variables
  
  # Print from index 0 to len(cir) - 1
  def printFullLinear(self): #Easy
    # To Do
    #pass # Remove this line

    for i in range(len(self.cir)):
      if i == len(self.cir)-1:
        print(self.cir[i])
      else:
        print(self.cir[i], end = ', ')
  
  # Print from start index and total size elements
  def printForward(self): #Easy
    # To Do
    #pass # Remove this line

    idx = self.start
    for i in range(self.size):
      if i == self.size-1:
        print(self.cir[idx])
      else:
        print(self.cir[idx], end = ', ')
      idx = (idx+1)%len(self.cir)
    #print(self.cir)

  
  def printBackward(self): #Easy
    # To Do
    #pass # Remove this line

    l_idx = (self.start+self.size-1)%len(self.cir)
    for i in range(self.size):
      if i == self.size-1:
        print(self.cir[l_idx])
      else:
        print(self.cir[l_idx], end=', ')
      l_idx -= 1
      if i < 0:
        i = len(self.cir)-1

  
  # With no null cells
  def linearize(self): #Medium
    # To Do
    #pass # Remove this line

    idx = self.start
    arr = [None]*self.size
    for i in range(self.size):
      arr[i] = self.cir[idx]
      idx = (idx+1)%len(self.cir)
    self.cir = arr
  
  # Do not change the Start index
  def resizeStartUnchanged(self, newcapacity): #Medium
    # To Do
    #pass # Remove this line

    idx = self.start
    idx1 = self.start
    n_cir = [None]*newcapacity
    for i in range(self.size):
      n_cir[idx1] = self.cir[idx]
      idx = (idx+1)%len(self.cir)
      idx1 = (idx1+1)%len(n_cir)
    self.cir = n_cir
  
  # This method will check whether the array is palindrome or not
  def palindromeCheck(self): #Hard
    # To Do
    #pass # Remove this line

    idx = self.start
    l_idx = (self.start+self.size-1)%len(self.cir)
    loopCount = self.size//2
    flag = True
    for i in range(loopCount):
      if self.cir[idx] == self.cir[l_idx]:
        flag = True
      else:
        flag = False
      idx = (idx+1)%len(self.cir)
      l_idx -= 1
      if l_idx < 0:
        l_idx = len(self.cir)-1
    if flag == True:
      print("This array is a palindrome")
    else:
      print("This array is NOT a palindrome")


  # This method will sort the values by keeping the start unchanged
  def sort(self):
    # To Do
    #pass # Remove this line

    idx = self.start
    for i in range(self.size):
      for j in range(self.size-i-1):
        #print(idx)
        if self.cir[idx] > self.cir[(idx+1)%len(self.cir)]:          
          temp = self.cir[idx]          
          self.cir[idx] = self.cir[(idx+1)%len(self.cir)]
          self.cir[(idx+1)%len(self.cir)] = temp
        idx = (idx+1)%len(self.cir)
      idx = self.start

  
  # This method will check the given array across the base array and if they are equivalent interms of values return true, or else return false
  def equivalent(self, cir_arr):
    # To Do
    #pass # Remove this line

    idx = self.start
    idx1 = cir_arr.start
    flag = True
    if self.size != cir_arr.size:
      flag = False
    else:
      for i in range(self.size):
        #for j in range(cir_arr.size):
          if self.cir[idx] == cir_arr.cir[idx1]:
            flag = True
          else:
            flag = False
            break
          idx = (idx+1)%len(self.cir)
          idx1 = (idx1+1)%len(cir_arr.cir)
    if flag == True:
      return "True"
    else:
      return "False"


  # the method take another circular array and returns a linear array containing the common elements between the two circular arrays.
  def intersection(self, c2):
    # To Do
    #pass # Remove this line

    idx = self.start
    idx1 = c2.start
    narr = []
    for i in range(self.size):
      idx1 = c2.start
      for j in range(c2.size):
        if self.cir[idx] == c2.cir[idx1]:
          narr += [c2.cir[idx1]]
        idx1 = (idx1+1)%len(c2.cir)
      idx = (idx+1)%len(self.cir)

    return narr
```
