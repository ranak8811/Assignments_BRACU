```python
# string task-1

wrd = input('Enter any word: ')

cu = 0
cl = 0

for i in wrd:
  if 'A'<=i<='Z':
    cu += 1
  elif 'a'<=i<='z':
    cl += 1

fu = ''
lu = ''

if cu>cl:
  for i in wrd:
    if 'A'<=i<='Z':
      fu += i
    elif 'a'<=i<='z':
      ab = ord(i)
      cd = chr(ab-32)
      fu += cd
  print(fu)

else:
  for i in wrd:
    if 'a'<=i<='z':
      lu += i
    elif 'A'<=i<='Z':
      ab = ord(i)
      cd = chr(ab+32)
      lu += cd
  print(lu)
```

```python
# string task-5

pas = 'OhMyBR@CU20'
uc = 0
lc = 0
dc = 0
sc = 0

for i in pas:
  if 'A'<=i<='Z':
    uc += 1
  elif 'a'<=i<='z':
    lc += 1
  elif '0'<=i<='9':
    dc += 1
  elif i=='_' or i=='$' or i=='#' or i=='@' :
    sc += 1

if uc>0 and lc>0 and dc>0 and sc>0 :
  print('OK')
else:
  if uc == 0:
    print('Uppercase missing')
  if lc == 0:
    print('Lowercase missing')
  if dc == 0:
    print('Digit missing')
  if sc == 0:
    print('Special missing')
```

```python
#dict task-1
d1 = {'a': 100, 'b': 100, 'c': 200, 'd': 300}
d2 = {'a': 300, 'b': 200, 'd': 400, 'e': 200}
nd = {}

for k,v in d1.items():
  nd[k] = v
for k,v in d2.items():
  if k in nd:
    nd[k] = d1[k]+v
  else:
    nd[k] = v

l = list(nd.values())
nl = []
for i in l:
  if i not in nl:
    nl.append(i)
nl.sort()
print(nd)
print('Values: ',tuple(nl))
```

```python
#dict task-2

import itertools
l = []
for i in itertools.count():
  ip = input('List elements: ')
  if ip == 'STOP':
    break
  else:
    l.append(int(ip))

d = {}
for num in l:
  if num in d:
    d[num] = d[num]+1
  else:
    d[num] = 1
print('Desired dictionary: ',d)

for k,v in d.items():
  print(f'{k} - {v} times')
```

```python
#func task-1
h = int(input('Enter your height in centimeters: '))
w = int(input('Enter your weight in kilograms: '))
def BMI_calc(height,weight):
  him = height/100
  BMI = weight/him**2

  if BMI<18.5:
    return BMI,'Underweight' 
  elif 18.5<=BMI<24.9:
    return BMI,'Normal'
  elif 25<=BMI<=30:
    return BMI,'Overweight'
  elif BMI>30:
    return BMI,'Obese'

res1,res2 = BMI_calc(h,w)
print(f"Score is {round(res1,1)}. You are {res2}")
```

```python
#func task-2
def summation(mi,ma,di):
  sum = 0
  for i in range(mi,ma,di):
    if i%di == 0:
      sum += i
  return sum 

summation(0,10,2)
```

```python
#func task-3
def buy_burger(name,location = 'Mohakhali'):
  d = {'BBQ Chicken Cheese Burger':250, 'Beef Burger':170, 'Naga Drums':200 }
  for i in d.keys():
    if i == name:
      am = d[i]
  tax = am*0.08
  amwt = am + tax
  if location == 'Mohakhali':
    tam = amwt + 40
  else:
    tam = amwt + 60
  return tam
  
buy_burger('Beef Burger')
```

```python
#func task-4
def domain_exchange(f_mail,new_d,old_d='gmail.com'):
  es = ''
  for i in range(len(f_mail)):
    if f_mail[i] == '@':
      p = i
      break
  wrd = f_mail[p+1::]
  if wrd == new_d:
    print('Unchanged: ',f_mail)
  else:
    b_wrd = f_mail[:p+1:]
    print(f'Changed: {b_wrd}{new_d}')

domain_exchange('bob@sheba.xyz', 'sheba.xyz')
```

```python
#func task-5
def palindrome(wrd):
  es = ''
  for i in wrd:
    if i == ' ':
      continue
    else:
      es += i
  n_wrd = es[-1:-len(es)-1:-1]
  if n_wrd == es:
    print('Palindrome')
  else:
    print('Not palindrome')

palindrome('nurses run')
```

```python
#func task-6
def capitalize(st1):
    st2 = ''
    for a in range(len(st1)):
        if a == 0:
            st2 += chr(ord(st1[a])-32)
        elif st1[a-2] == "." or st1[a-2] == "!" or st1[a-2] == "?":
            st2 += chr(ord(st1[a])-32)
        elif st1[a] == "i" and st1[a-1] == " " and st1[a+1] == " ":
            st2 += chr(ord(st1[a]) - 32)
        else:
            st2 += st1[a]
    return st2

#string = input('Enter a string: ')
string = 'my favourite animal is a dog. a dog has sharp teeth so that it can eat flesh very easily.\
do you know my pet dog’s name? i love my pet very much.'
print(capitalize(string))
```
