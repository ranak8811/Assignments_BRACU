```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  #  Design the constructor based on data type of a. If 'a' is built in python list then
  #  Creates a linked list using the values from the given array.
    self.head = Node(None, None, None)
    # To Do
    #pass # Remove this line

    if type(a) == list:
      DH = self.head
      tail = DH
      for i in range(len(a)):
        n = Node(a[i], None, None)
        tail.next = n
        n.prev = tail
        tail = tail.next
        tail.next = DH
        DH.prev = tail
    else:
      self.head = a


  
  # Counts the number of Nodes in the list
  def countNode(self):
    # To Do
    #pass # Remove this line

    temp = self.head
    c = 0
    while temp.next != self.head:
      c += 1
      temp = temp.next
    return c


  
  # prints the elements in the list
  def forwardprint(self):
    # To Do
    #pass # Remove this line


    temp = self.head.next
    while temp != self.head:
      if temp.next == self.head:
        print(temp.element, end = '.')
      else:
        print(temp.element, end = ', ')
      temp = temp.next
    print()


  # prints the elements in the list backward
  def backwardprint(self):
    # To Do
    #pass # Remove this line

    temp = self.head.prev
    while temp != self.head:
      if temp.prev == self.head:
        print(temp.element, end = '.')
      else:
        print(temp.element, end = ', ')
      temp = temp.prev
    print()


  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    #pass # Remove this line

    if (idx < 0) or (idx > self.countNode()):
      print ("Index Error")
      return None

    temp = self.head.next
    c = 0
    while temp != self.head:
      if c == idx:
        return temp
      else:
        c += 1
        temp = temp.next
    return 'index error'


  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    #pass # Remove this line

    temp = self.head.next
    c = 0
    while temp != self.head:
      if temp.element == elem:
        return c
      else:
        c += 1
        temp = temp.next
    return -1


  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    # To Do
    #pass # Remove this line


     if (idx < 0) or (idx > self.countNode()):
      print ("Index Error")
      return -1
     temp = self.head.next
     c = 0

     while temp != self.head:
        if c == idx:
          temp = temp
          break
        else:
          c += 1
          temp = temp.next 
    #  newNode = Node(elem, None, None)
    #  pred = temp.prev
    #  newNode.next = temp
    #  temp.prev = newNode
    #  pred.next = newNode
    #  newNode.prev = pred
     newNode = Node(elem, temp, temp.prev) 
     pred = temp.prev
     pred.next = newNode
     temp.prev = newNode



  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    #pass # Remove this line


    if (idx < 0) or (idx > self.countNode()):
      print ("Index Error")
      return -1

    temp = self.head.next
    c = 0

    while temp != self.head:
      if c == idx:
        temp = temp
        break
      else:
        c += 1
        temp = temp.next 

    pred = temp.prev
    after = temp.next
    pred.next = after
    after.prev = pred
    return str(temp.element)
```
